```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Создаем базу данных
conn = sqlite3.connect('salarydatabase.db')
cursor = conn.cursor()

# Создание таблицы "salarydatabase" и вводим переменные
cursor.execute('''CREATE TABLE IF NOT EXISTS salarydatabase (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT NOT NULL,
    Salary_1month INTEGER NOT NULL,
    Salary_2month INTEGER NOT NULL,
    Salary_3month INTEGER NOT NULL,
    Bonus_everymonth INTEGER NOT NULL,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
)''')

# загружаем данные из датасета в таблицу, указываем, что таблица постоянно обновляется, праймри кей не отображаем при выводе
df = pd.read_csv('salary_dataset.csv')
df.to_sql('salarydatabase', conn, if_exists='replace', index=False)

#Делаем первый запрос: выводим всю таблицу
print (f"Выводим всю таблицу")
cursor.execute("SELECT * FROM salarydatabase")
salarydatabase = cursor.fetchall()
for Name in salarydatabase:
    print(Name)

#Считаем количество записей в таблице
print (f"Количество записей в таблице:")
cursor.execute('SELECT count(*) FROM salarydatabase')
num_rows = cursor.fetchone()[0]
print(num_rows)

#Выводим только тех, у кого ЗП за 1 месяц =>1000, но меньше либо разва 2200 за второй
print(f"Сотрудники, у которых зарплата за первый месяц больше либо равна 1000, а за второй месяц меньше либо равна 2200:")
cursor.execute('SELECT Name, Salary_1month, Salary_2month FROM salarydatabase WHERE Salary_1month>=1000 and Salary_2month<=2200')
salarydatabase = cursor.fetchall()
for Name in salarydatabase:
    print(Name)

# Выводим только тех, у кого ЗП за 2 месяц между 1500 и 2000
print(f"Сотрудники, у которых зарплата второго месяца находится в диапазоне от 1500 до 2000")
cursor.execute('SELECT Name, Salary_2month FROM salarydatabase WHERE Salary_2month between 1500 and 2000')
salarydatabase = cursor.fetchall()
for Name in salarydatabase:
    print(Name)

# Находим, сколько компания заплатила за каждый месяц работы
print(f"Сумма потраченная компанией за каждый месяц")
cursor.execute('SELECT sum(Salary_1month+Bonus_everymonth), sum(Salary_2month+Bonus_everymonth), sum(Salary_3month+Bonus_everymonth) FROM salarydatabase')
salarydatabase = cursor.fetchall()
for Name in salarydatabase:
    print(Name)

#Находим среднее значение по каждому месяцу
print(f"Среднее значение зарплаты за каждый месяц")
cursor.execute('SELECT AVG(Salary_1month+Bonus_everymonth), AVG(Salary_2month+Bonus_everymonth), AVG(Salary_3month+Bonus_everymonth) FROM salarydatabase')
salarydatabase = cursor.fetchall()
for Name in salarydatabase:
    print(Name)

# Топ 3 сотрудника по средней ЗП за 3 месяа
print(f"ТОП 3 сотрудника по срдней ЗП за 3 месяца")
cursor.execute('SELECT Name, (Salary_1month + Salary_2month + Salary_3month + Bonus_everymonth*3)/3 as "Salary" FROM salarydatabase ORDER BY "Salaty" DESC Limit 3')
salarydatabase = cursor.fetchall()
for Name in salarydatabase:
    print(Name)

#Начинаме подготовку данных для графика. Преобразуем столбцы с зарплатами в строки
df_long = pd.melt(
    df,
    id_vars=["Name", "Bonus_everymonth"],
    value_vars=["Salary_1month", "Salary_2month", "Salary_3month"],
    var_name="Месяц",
    value_name="Зарплата"
)

# Переименуем месяцы
month_map = {
    "Salary_1month": "Январь",
    "Salary_2month": "Февраль",
    "Salary_3month": "Март"
}
df_long["Месяц"] = df_long["Месяц"].map(month_map)

# Упорядочим месяцы
df_long["Месяц"] = pd.Categorical(
    df_long["Месяц"],
    categories=["Январь", "Февраль", "Март"],
    ordered=True
)

# Добавим зарплату с премией
df_long["Зарплата_с_премией"] = df_long["Зарплата"] + df_long["Bonus_everymonth"]
# формируем график
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=df_long,
    x="Месяц",
    y="Зарплата_с_премией",
    hue="Name",
    marker="o",
    linewidth=2
)
plt.title("Динамика зарплат с премией")
plt.ylabel("Зарплата (руб)")
plt.grid(True)
plt.legend(title="Сотрудник")
plt.show()


# Закрытие соединения и сохранение изменений
conn.commit()
conn.close()
```
